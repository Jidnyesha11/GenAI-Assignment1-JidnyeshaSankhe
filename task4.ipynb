```python
# Product List
products = [
    {"name": "Laptop", "category": "Electronics"},
    {"name": "Mouse", "category": "Electronics"},
    {"name": "Chair", "category": "Furniture"},
    {"name": "Table", "category": "Furniture"},
    {"name": "Pen", "category": "Stationery"}
]

# Price Dictionary
price_dict = {
    "Laptop": 50000,
    "Mouse": 800,
    "Chair": 3000,
    "Table": 6000,
    "Pen": 20
}

#Create catalog
catalog=[]

for product in products:
    name=product["name"]
    price=price_dict[name]
    catalog.append((name, price))

print("Catalog:")
print(catalog)
```

```python
# create category_to_products dictionary
category_to_products={}

for product in products:
    category=product["category"]
    name=product["name"]

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(name)

print("\nCategory to Products:")
print(category_to_products)
```

```python
# category with max products
max_category = ""
max_count = 0

for category in category_to_products:
    count = len(category_to_products[category])

    if count > max_count:
        max_count = count
        max_category = category

print("\nCategory with Maximum Products:")
print(max_category)
```
