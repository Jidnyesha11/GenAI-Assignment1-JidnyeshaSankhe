```python
#create dictionary

price_dict = {
    "Laptop":50000,
    "Mobile":25000,
    "Keyboard":1200,
    "Mouse":700,
    "Monitor":12000,
    "Speaker":3500
}

print(price_dict)
```

```python
# add  product
price_dict["Printer"] = 8000

print(price_dict)
```

```python
# update the existing product price

product=input("Enter product to update:")

if product in price_dict:
    new_price=int(input("Enter new price:"))
    price_dict[product] = new_price
    print("Price Updated")
else:
    print("Product Not Found")

print(price_dict)
```

```python
# remove product using del 

product=input("Enter product to remove:")

if product in price_dict:
    del price_dict[product]
    print("Product Removed")
else:
    print("Product Not Found")

print(price_dict)
```

```python
# print avg price

total = sum(price_dict.values())

average = total / len(price_dict)

print("Average Price =", average)
```

```python
# minimum and maximum price

maximum = max(price_dict.values())
minimum = min(price_dict.values())

print("Maximum Price:", maximum)
print("Minimum Price:", minimum)
```
